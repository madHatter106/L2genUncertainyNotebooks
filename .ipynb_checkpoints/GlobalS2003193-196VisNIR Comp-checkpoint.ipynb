### GOAL: plot vis/nir MC sim results for global S2003193-196_4D scene

```python
# IMPORT STUFF
from obpgpyutils.LxReader.readers import CL3bReader
from matplotlib.pylab import rcParams
from matplotlib import rc
from matplotlib.ticker import LogLocator
import matplotlib.pyplot as pl
import os
import numpy as np
import seaborn as sb
```

```python
# WHERE TO PLOT
% matplotlib inline
```

```python
# GRAPHICS SET-UP
rc('font', **{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=False)
sb.set_style('darkgrid')
mostFontSize = 16
rcParams['xtick.labelsize'] = mostFontSize
rcParams['ytick.labelsize'] = mostFontSize
rcParams['xtick.minor.visible'] = True
rcParams['xtick.minor.size'] = 5
rcParams['xtick.minor.width'] = 2
rcParams['xtick.direction'] = 'Down'
```

```python
#PATHS
maindir = '/accounts/ekarakoy/disk02/UNCERTAINTIES/Monte-Carlo/SWF/DATA/L2_L3/GLOBAL'
stdir = 'StandardRun'
visdir = 'VisOnly'
stfname = 'S20031932003196.L3b_4D_SU4.nc'
visfname = 'S20031932003196.L3b_4D_SU5.nc'
STPATH = os.path.join(maindir,stdir,stfname)
VISPATH = os.path.join(maindir,visdir,visfname)
print("Found %s: %s" % (stfname,os.path.exists(STPATH)))
print("Found %s: %s" % (visfname,os.path.exists(VISPATH)))
mainSaveDir = '/accounts/ekarakoy/DEV-ALL/TestingGround/PLOTTING/Comparisons'
```

```python
# OTHER GLOBALS
BANDS = [412,443,490,510,555,670]
PRODLIST = ['Rrs_unc_%d' % band for band in BANDS] + ['Rrs_%d' % band for band in BANDS]
```

```python
# LOAD
with CL3bReader(STPATH,products=PRODLIST) as stl3:
    pass


with CL3bReader(VISPATH,products=PRODLIST) as visl3:
    pass
```

```python
def PlotComparisonHist(data1, data2,**kwargs):
    name1 = kwargs.pop('name1', None)
    name2 = kwargs.pop('name2', None)
    rng = kwargs.pop('rng', None)
    ax = kwargs.pop('ax', None)
    normGraph = kwargs.pop('normalize',False)
    saveName = kwargs.pop('savename', None)
    yticks = kwargs.pop('yticks', None)
    ant = kwargs.pop('annotateThis', None)
    #if not rng:
    #    rng = (min(data1.min(), data2.min()), max(data1.max(), data2.max()))
    rngMin,rngMax = min(data1.min(),data2.min()),max(data1.max(),data2.max())
    #databins = np.logspace(np.log10(rng[0]),np.log10(rng[1]),100)
    databins = np.logspace(np.log10(rngMin),np.log10(rngMax),100)
    rcParams['xtick.minor.size'] = 5
    ax.hist(data1, bins=databins, histtype='stepfilled', label=name1,
           normed=normGraph)
    ax.hist(data2, bins=databins,histtype='stepfilled', label=name2,
           color='gray', alpha=0.7,normed=normGraph)
    ax.hist(data1, bins=databins, color='navy',normed=normGraph, linewidth=1,
           histtype='step')
    ax.hist(data2, bins=databins,color='k',normed=normGraph,linewidth=1,
           histtype='step')
    ax.set_xscale('log')
    ax.legend(loc='best',fontsize=12)
    ax.set_xlim(rng)
    minorLocator = LogLocator(subs=np.arange(2,11)/10)
    ax.xaxis.grid(True, which='minor')
    if yticks is not None:
        ax.yaxis.set_ticks(yticks)
    ax.xaxis.set_minor_locator(minorLocator)
    ax.spines['top'].set_visible(False)
    ax.xaxis.set_ticks_position('bottom')
    ax.ticklabel_format(axis='y', style='sci',scilimits=(-1,3))
    if ant:
        ax.annotate(ant['msg'],xy=ant['xyCoor'],fontsize=ant['fontSize'])
```

```python
# Plotting Absolute
fig,axs = pl.subplots(ncols=2, nrows=3, figsize=(12,12))
rngDict = {412:(5e-5,8e-4), 443:(4e-5,8e-4), 490:(3e-5,5e-4),
           510:(2e-5,5e-4), 555:(2e-5,4e-4), 670:(9.5e-6,2e-4)}
yticDict = {412:np.arange(0,1.7e5,5e4), 443:np.arange(0,2.1e5,4e4), 
            490:np.arange(0,2.6e5,5e4), 510:np.arange(0,2.6e5,5e4),
            555:np.arange(0,3.1e5,5e4), 670:np.arange(0,2.1e5,4e4)}
txtxyDict = {412:(3e-4,1e5),443:(3e-4,1.2e5),490:(2e-4,1.5e5),
           510:(2e-4,1.5e5),555:(1.7e-4,2e5),670:(7e-5,1.2e5)}
for band,ax in zip(BANDS, axs.ravel()):
    annotationDict = {'msg': r'$\lambda=%dnm$' % band, 'xyCoor':(txtxyDict[band]),
                      'fontSize': 12}
    PlotComparisonHist(stl3.productDict['Rrs_unc_%d' % band],
                       visl3.productDict['Rrs_unc_%d' % band],
                       name1='VIS + NIR',name2='VIS',ax=ax,rng=rngDict[band],
                       yticks=yticDict[band],annotateThis=annotationDict)
    if band in [412,490,555]:
        ax.set_ylabel('bin count', fontsize=mostFontSize)
    if band in [555,670]:
        ax.set_xlabel(r'$Rrs$ uncertainty ($sr^{-1}$)', fontsize=mostFontSize)
```

```python
fig.savefig(os.path.join(mainSaveDir,'hist_absolute.pdf'),dpi=300,format='pdf')
```

```python
# Plotting Relative
fig2,axs2 = pl.subplots(ncols=2, nrows=3, figsize=(12,12))
rngDict = {412:(2e-1,5e1), 443:(2e-1,5e1), 490:(2e-1,2e1),
           510:(2e-1,3e1), 555:(3e-1,4e1), 670:(5e-1,4e2)}
yticDict = {412:np.arange(0,1.3e5,4e4), 443:np.arange(0,1.3e5,4e4), 
            490:np.arange(0,1.3e5,4e4), 510:np.arange(0,2.6e5,5e4),
            555:np.arange(0,1.3e5,4e4), 670:np.arange(0,1.3e5,3e4)}
txtxyDict = {412:(1e1,8e4),443:(1e1,.8e5),490:(5,.8e5),
           510:(6,1.5e5),555:(1e1,.8e5),670:(6e1,.8e5)}
for band,ax2 in zip(BANDS, axs2.ravel()):
    annotationDict = {'msg': r'$\lambda=%dnm$' % band, 'xyCoor':(txtxyDict[band]),
                      'fontSize': 12}
    stIdx = np.argwhere(stl3.productDict['Rrs_%d' % band] != 0)
    visIdx = np.argwhere(visl3.productDict['Rrs_%d' % band] != 0)
    data1 = stl3.productDict['Rrs_unc_%d' % band][stIdx] /\
            stl3.productDict['Rrs_%d' % band][stIdx] * 100
    data2 = visl3.productDict['Rrs_unc_%d' % band][visIdx] /\
            visl3.productDict['Rrs_%d' % band][visIdx] * 100
    PlotComparisonHist(data1,data2, name1='VIS + NIR', name2='VIS',
                       ax=ax2,rng=rngDict[band], yticks=yticDict[band],
                      annotateThis=annotationDict)
    if band in [412,490,555]:
        ax2.set_ylabel('bin count', fontsize=mostFontSize)
    if band in [555,670]:
        ax2.set_xlabel(r'$Rrs$ uncertainty ($\%$)', fontsize=mostFontSize)
```

```python
fig2.savefig(os.path.join(mainSaveDir,'hist_relative.pdf'),dpi=300,format='pdf')
```

```python
# Get Regional masks
stl3.GetRegionMasks(['olig','eutr'])
visl3.GetRegionMasks(['olig','eutr'])
```

```python
with open(os.path.join(mainSaveDir,'Medians.txt'),mode='w') as f:
    print("Medians (1/sr) - All regions:", file=f)
    for band in BANDS:
        stAbsMedian = np.median(stl3.productDict['Rrs_unc_%d' % band])
        visAbsMedian = np.median(visl3.productDict['Rrs_unc_%d' % band])
        print("%d -- VIS+NIR: %.2e -- VIS: %.2e" % (band, stAbsMedian, visAbsMedian),
             file=f)
    print('------------------------------', file=f)
    print("Medians (1/sr) - Oligotrophic:", file=f)
    for band in BANDS:
        stAbsMedian = np.median(stl3.productDict['Rrs_unc_%d' % band][stl3.regionMaskDict['olig']])
        visAbsMedian = np.median(visl3.productDict['Rrs_unc_%d' % band][visl3.regionMaskDict['olig']])
        print("%d -- VIS+NIR: %.2e -- VIS: %.2e" % (band, stAbsMedian, visAbsMedian),
             file=f)
    print('------------------------------', file=f)
    print("Medians (1/sr) - Eutrophic:", file=f)
    for band in BANDS:
        stAbsMedian = np.median(stl3.productDict['Rrs_unc_%d' % band][stl3.regionMaskDict['eutr']])
        visAbsMedian = np.median(visl3.productDict['Rrs_unc_%d' % band][visl3.regionMaskDict['eutr']])
        print("%d -- VIS+NIR: %.2e -- VIS: %.2e" % (band, stAbsMedian, visAbsMedian),
             file=f)
    print('------------------------------', file=f)
    print('------------------------------', file=f)
    print("Medians (%) - All regions:", file=f)
    for band in BANDS:
        stIdx = np.argwhere(stl3.productDict['Rrs_%d' % band] != 0)
        visIdx = np.argwhere(visl3.productDict['Rrs_%d' % band] != 0)
        data1 = stl3.productDict['Rrs_unc_%d' % band][stIdx] /\
                stl3.productDict['Rrs_%d' % band][stIdx] * 100
        data2 = visl3.productDict['Rrs_unc_%d' % band][visIdx] /\
                visl3.productDict['Rrs_%d' % band][visIdx] * 100
        stRelMedian = np.median(data1)
        visRelMedian = np.median(data2)
        print("%d -- VIS+NIR: %.2e -- VIS: %.2e" % (band, stRelMedian, visRelMedian), file=f)
    print('------------------------------', file=f)
    print("Medians (%) - Oligotrophic:", file=f)
    for band in BANDS:
        data1 = stl3.productDict['Rrs_unc_%d' % band][stl3.regionMaskDict['olig']] /\
                stl3.productDict['Rrs_%d' % band][stl3.regionMaskDict['olig']] * 100
        data2 = visl3.productDict['Rrs_unc_%d' % band][visl3.regionMaskDict['olig']] /\
                visl3.productDict['Rrs_%d' % band][visl3.regionMaskDict['olig']] * 100
        stIdx = np.argwhere(data1 != 0)
        visIdx = np.argwhere(data2 != 0)
        stRelMedian = np.median(data1[stIdx])
        visRelMedian = np.median(data2[visIdx])
        print("%d -- VIS+NIR: %.2e -- VIS: %.2e" % (band, stRelMedian, visRelMedian), file=f)
    print('------------------------------', file=f)
    print("Medians (%) - Eutrophic:", file=f)
    for band in BANDS:
        data1 = stl3.productDict['Rrs_unc_%d' % band][stl3.regionMaskDict['eutr']] /\
                stl3.productDict['Rrs_%d' % band][stl3.regionMaskDict['eutr']] * 100
        data2 = visl3.productDict['Rrs_unc_%d' % band][visl3.regionMaskDict['eutr']] /\
                visl3.productDict['Rrs_%d' % band][visl3.regionMaskDict['eutr']] * 100
        stIdx = np.argwhere(data1 != 0)
        visIdx = np.argwhere(data2 != 0)
        stRelMedian = np.median(data1[stIdx])
        visRelMedian = np.median(data2[visIdx])
        print("%d -- VIS+NIR: %.2e -- VIS: %.2e" % (band, stRelMedian, visRelMedian), file=f)
```
