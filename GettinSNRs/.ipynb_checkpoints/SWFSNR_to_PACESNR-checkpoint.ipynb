```python
def ApplyPolyFit(cf,lt,deg=4):
    y = 0
    for i in range(deg):
        y += cf[i] * (lt** (deg - i))
    y += cf[deg]
    return y
```

```python
def SWF2PACEFacs(idstr,verbose=True):
    "Return noise scale factors from SWF to PACE for given SNR set"
    bands=[412,443,490,510,555,665,748,865]
    ltyps=dict.fromkeys(bands)
    targSnr = dict.fromkeys(bands)
    targSnr=dict(dfGsdSnr.loc[bands,idstr])
    ltyps = dict(dfGsdSnr.loc[bands].Ltyp)
    if verbose:
        print(targSnr)
    reqFac=dict.fromkeys(bands)
    targFac=dict.fromkeys(bands)
    reqSnr=[1000]*6 + [600]*2
    for i,band in enumerate(bands):
        if band == 665:
            cband = '670'
        elif band == 748:
            cband ='765'
        else:
            cband = str(band)
        oldsnr=ApplyPolyFit(coefDict[cband],ltyps[band])
        reqFac[band] = reqSnr[i]/oldsnr
        targFac[band] = targSnr[band]/oldsnr
        if verbose:
            print("%s=>old snr: %f - factor(req): %f fac(bsl): %f" % (band,oldsnr,
                                                                  reqFac[band],targFac[band]))
    return targFac
```

```python
def WriteParfile(idstr,bands,targFac):
    """ writes parfile for given noise scale factors"""
    fname='../../parfiles/PACE/parfile_noisy_' + idstr
    nsList = '['
    for band in bands:
        band = int(band)  
        if band == '-1':
            nsList += '%s' % band
        else:
            if band == 670:
                band = 665
            elif band == 765:
                band = 748
            nsList+='%.6f' % targFac[band]
        if band != int(bands[-1]):
            nsList +=','
    nsList +=']'
    with open(fname,'w') as f:
        f.write('l2prod="Rrs_nnn,a_nnn_qaa,b_nnn_qaa,Lt_nnn,chlor_a"\n')
        f.write('add_noise_sigma=1\n')
        f.write('noise_scale='+str(nsList)+'\n')
        f.write('oformat=netcdf4')
```
