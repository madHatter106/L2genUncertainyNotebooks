```python
def ApplyPolyFit(cf,lt,deg=4):
    y = 0
    for i in range(deg):
        y += cf[i] * (lt** (deg - i))
    y += cf[deg]
    return y
```

```python
def SWF2PACEFacs(df_targ, **kwargs):
    "Return noise scale factors from SWF to PACE for given SNR set"
    
    mcbands = kwargs.pop('mcbands', [412, 443, 490, 510, 555, 670, 765, 865])
    verbose = kwargs.pop('verbose', True)
    pacebands = kwargs.pop('pacebands',[412, 443, 490, 510, 555, 665, 748, 865])
    ltypDict = dfGsdSnr.loc[pacebands,'Ltyp'].to_dict()
    targFac = dict.fromkeys(mcbands)
    print("creating sim params for:")
    print(df_targ)
    for mcband, paceband in zip(mcbands,pacebands):
        print(mcband, paceband)
        oldsnr = ApplyPolyFit(coefDict[str(mcband)], ltypDict[paceband])
        targFac[mcband] = df_targ.loc[mcband].SNR / oldsnr
        if verbose:
            print('-' * 80)
            print("%s=>old snr: %f - new snr: %f factor: %f" % 
                  (mcband, oldsnr, df_targ.loc[mcband].SNR, targFac[mcband]))
    return targFac
```
