```python
import matplotlib.pyplot as pl
from netCDF4 import Dataset as DS
import numpy as np
import seaborn as sb
from tqdm import tqdm_notebook
import glob
import os
import re
from matplotlib import rcParams
```

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container {width:90% !important;}</style>"))
```

```python
rcParams['axes.formatter.limits'] = (-2,2)
rcParams['xtick.labelsize'] = 16
rcParams['ytick.labelsize'] = 16
sb.set_style(rc=rcParams)
%matplotlib inline
```

```python

bands = [412, 443, 490, 510, 555, 670]
def GetL3mProds(path, verbose=True, getLatLon=False):
    """Loads data into dictionaries for Lt and Rrs."""
    ltDict = dict.fromkeys(bands)
    rrsDict = dict.fromkeys(bands[:-2])
    with DS(path) as ds:
        varbls = ds.variables
        for band in bands:
            try:
                ltDict[band] = varbls['Lt_%d' % band][:]
                rrsDict[band] = varbls['Rrs_%d' % band][:]
            except KeyError as e:
                if verbose:
                    print(e, ' ... skipping')
                continue
        if getLatLon:
            geoDict = {k: varbls[k][:] for k in ['lat', 'lon']}
            return ltDict, rrsDict, geoDict
    return ltDict, rrsDict

def GetDiff(inpDict, j, bsl='baseline', delta=False, rhow=False):
    factor = 1.0
    if rhow:
        factor = np.pi
    dDict = dict.fromkeys(bands)
    for band in bands:
        rrs = inpDict[j]['rrs'][band]
        rrsb = inpDict[bsl]['rrs'][band]
        if delta:
            diff = (rrs - rrsb) / (rrs + 1e-6)
        else:
            diff = (rrs - rrsb) * factor
        dDict[band] = diff.compressed()
    return dDict

def GetDeltaDiff(inpDict, j, bsl='baseline', rhow=False):
    """
        Return band-keyed dict with dRrs/dLt.
    """
    factor = 1.0
    if rhow:
        factor = np.pi
    dDict = dict.fromkeys(bands)
    for band in bands:
        rrs = inpDict[j]['rrs'][band]
        rrsb = inpDict[bsl]['rrs'][band]
        lt = inpDict[j]['lt'][band]
        ltb = inpDict[bsl]['lt'][band]
        deltaDiff = (rrs - rrsb) / (lt - ltb) * factor
        dDict[band] = deltaDiff.compressed()
    return dDict


def PlotHist(data, ax, **kwargs):
    bins = kwargs.pop("bins",100)
    rng = kwargs.pop("rng", None)
    lbl = kwargs.pop("lbl", None)
    alpha = kwargs.pop("alpha", 0.5)
    col = kwargs.pop("col", 'k')
    logOn = kwargs.pop('log', False)
    ax.hist(data,bins=bins, range=rng, normed=True, label=lbl, alpha=alpha,
           histtype="stepfilled", log=logOn, color=col)
    ax.hist(data, bins=bins, range=rng, normed=True, histtype="step", linewidth=3,
            color="k")
    ax.set_xlim(rng)
    ax.get_xaxis().set_tick_params(labelsize=18)
    ax.get_yaxis().set_tick_params(labelsize=18)

def DrawPlayGround(ax, **kwargs):
    title = kwargs.pop('title', None)
    subRedThresh = kwargs.pop('subRedThresh', 1e-3)
    redThresh = kwargs.pop('redThresh', 2e-3)
    axYmax = kwargs.pop('ymax', 2e-3)
    ax.set_xlim(412,700)
    ax.set_ylim(0,axYmax)
    ax.set_xticks([int(band) for band in bands])
    ax.set_title(title, fontsize=22)
    ax.set_ylabel(r'random error ($\rho$)', fontsize=17)
    ax.set_xlabel(r'$\lambda (nm)$', fontsize=18)
    ax.axvspan(412, 600, ymin=0, ymax=subRedThresh/axYmax,
               facecolor='gray', alpha=0.5)
    ax.axvspan(600, 700, ymax=redThresh/axYmax,
               facecolor='gray', alpha=0.5, label='threshold')

def MakeLinePlot(data, ax, ylim, idx=(1,10), pct=0.9, factor=1, **kwargs):
    ylbl = kwargs.pop("ylbl", None)
    xlbl = kwargs.pop("xlbl", None)
    dataPts = {i: [] for i in range(*idx)}
    for k in range(*idx):
        col = colDict['%d' % k]
        for band in bands:
            arr = np.sort(data['%d' % k][band] * factor)
            dataPts[k].append(arr[:round(pct * arr.size)].max())
        ax.plot(bands, dataPts[k], color=col, marker='o', label='0.%d%%' % k, lw=2)
    ax.ticklabel_format(style='plain', axis='x')
    ax.set_ylim(ylim)
    if xlbl:
        ax.set_xlabel(xlbl, fontsize=20)
    if ylbl:
        ax.set_ylabel(ylbl, fontsize=20)
    ax.get_xaxis().set_tick_params(labelsize=18)
    ax.get_yaxis().set_tick_params(labelsize=18)
    return ax
```

```python
dataDir = '/accounts/ekarakoy/disk02/UNCERTAINTIES/Systematic_Error/PACE_SWF/All4Ds/'
fileList = glob.glob(os.path.join(dataDir, '*.L3m*'))
pat = re.compile(r'_PSES([0-9a-z]+)')
```

```python
fileList
```

```python
dataDict = {}
for file in tqdm_notebook(fileList):
    ext = str(int(pat.findall(file)[0]))
    if ext == '0':
        ext = 'baseline'
    dataDict[ext] = dict.fromkeys(['lt', 'rrs'])
    dataDict[ext]['lt'], dataDict[ext]['rrs'] = GetL3mProds(file)
```

```python
difDict = dict.fromkeys([str(i) for i in range(1,10)])
for k in difDict.keys():
    #difDict[k] = GetDiff(dataDict,k)
    print(k,dataDict[k]['lt'][412].shape, dataDict[k]['rrs'][412].shape)
```

```python
deltaDifDict = dict.fromkeys([str(i) for i in range(1,10)])
difDict = dict.fromkeys([str(i) for i in range(1, 10)])
for k in tqdm_notebook(difDict.keys()):
    difDict[k] = GetDiff(dataDict,k,delta=False,rhow=True)
    deltaDifDict[k] = GetDiff(dataDict,k,delta=True,rhow=True)
```

```python
keys = difDict.keys()
colDict = {k:c for k,c in zip(keys, sb.color_palette('husl', n_colors=len(keys)))}
colDict['1'] = sb.xkcd_rgb["dark blue"]
colDict['9'] = sb.xkcd_rgb["black"]
colDict['3'] = sb.xkcd_rgb["orange"]
colDict['7'] = sb.xkcd_rgb["green"]
colDict['4'] = sb.xkcd_rgb["brown"]
colDict['6'] = sb.xkcd_rgb["plum"]
```

```python
sb.set(font_scale=1.5)
f, axs=pl.subplots(ncols=2,nrows=3,figsize=(24,24))
bandRng = {412: (0, 25), 443: (0, 16), 490: (0, 12),
           510: (0, 12.5), 555: (0, 16), 670: (0, 50)}
f.suptitle('Relative Systematic Error, $90^{th}$ Percentile', fontsize=24, y=.93)
for band,ax in zip(bands, axs.ravel()):
    for k in range(1,10) :
        col = colDict['%d' % k]
        array = np.sort(deltaDifDict['%d' % k][band]*100)
        array = array[:round(0.9 * array.size)]
        label = r'0.%d%%: $\Delta_{max,90pct}=%.1f%%$' %(k, array.max())
        alph = 1 - k * 0.07
        PlotHist(array, ax=ax, alpha=alph, rng=bandRng[band], lbl=label, 
                col=col)
        ax.set_title(band, fontsize=24)
        ax.legend(fontsize=18)
    if band in [412,490,555]:
        ax.set_ylabel('freq', fontsize=20)
    if band in [555, 670]:
        ax.set_xlabel(r'$\frac{\Delta\rho}{\rho_{bsl}} \times 100\%$', fontsize=24)

f.savefig('/accounts/ekarakoy/Desktop/SysErr_rel.png',dpi=300, format='png')
```
