### Synopsis:

Here I analyze the results of 8 + 1 runs, where each of the 8 runs had $L_{TOA}$ perturbed by 0.1 and 0.5%. The goal is to confirm the impact of such a perturbation on the band of interest observed  in a previous all-band perturbation of similar level. A follow-up goal is to determine to what extent, if at all, a single-band perturbation affects any of the other bands. Of particular interest is the impact of NIR band perturbation.

```python
# generate regex matching pattern for 4D files
pert4DFilePattern = re.compile('(S[0-9]+.L3m_4D.+)')
```

```python
# load dictionary with relative paths to 4D data files, including baseline data
for k in relPathsDict.keys():
    if k == 'baseline':
        path = os.path.join(mainDir, bslDataDir, '*')
    else:
        path = os.path.join(mainDir, pertDataDir, k, '*')
    for l in glob.iglob(path):
        fileMatch = pert4DFilePattern.findall(l)
        if fileMatch:
            relPathsDict[k] = os.path.join(re.findall('(B[0-9a-zA-Z\W_]+)\*',
                                                      path)[0], fileMatch[0])
```

#### Loading each file and computing difference with baseline
* in Lt as sanity check,
* in Rrs for actual impact assessment

The important data structures here are two dictionaries, relPathsDict and dataDict:
* relPathsDict contains relative paths to the 4D data files
* dataDict uses same keys as relPathsDict, contains:
    * difference in Lt relative to baseline (sanity check) ['Lt']
    * difference in Rrs relative to baseline (impact assessment) ['Rrs']
    * the "baseline" entry contains baseline data

```python
bands = [412,443,490,510,555,670,765,865]
def GetL3mProds(path, verbose=True, getLatLon=False):
    """Loads data into dictionaries for Lt and Rrs."""
    ltDict = dict.fromkeys(bands)
    rrsDict = dict.fromkeys(bands[:-2])
    with DS(path) as ds:
        varbls = ds.variables
        for band in bands:
            try:
                ltDict[band] = varbls['Lt_%d' % band][:]
                rrsDict[band] = varbls['Rrs_%d' % band][:]
            except KeyError as e:
                if verbose:
                    print(e, ' ... skipping')
                continue
        if getLatLon:
            geoDict = {k: varbls[k][:] for k in ['lat', 'lon']}
            return ltDict, rrsDict, geoDict
    return ltDict, rrsDict

def ComputeDiffs(data1, bslData, computeRel=True):
    absdiff = data1 - bslData
    if computeRel:
        reldiff = absdiff / bslData
        return reldiff
    return absdiff

def GetDiffs(path, bslDict, computeRel=True):
    #common_bins = bslDict[]
    if bslDict:
        ltDct, rrsDct = GetL3mProds(path, verbose=False)
        ltDiffDct = dict.fromkeys(ltDct.keys())
        rrsDiffDct = dict.fromkeys(rrsDct.keys())
        for band in bands:
            ltDiffDct[band] = ComputeDiffs(ltDct[band], bslDict['Lt'][band], computeRel)
            try:
                rrsDiffDct[band] = ComputeDiffs(rrsDct[band], bslDict['Rrs'][band],
                                                computeRel)
            except KeyError:
                continue
        return ltDiffDct, rrsDiffDct
    else:
        print("no baseline data!")
        return None

def PlotData(dataDict, bandNum, rngDict, ylbl="Freq", relative=True,
             xlbl="$Wm^{-2}\mu m^{-1} sr^{-1}$"):
    f,axs = pl.subplots(ncols=2, nrows=3, figsize=(12,16))
    factor = 1
    if relative:
        factor = 100
    for band, ax in zip(bands[:-2], axs.ravel()):
        
        data = dataDict[band].compressed() * factor
        ax.hist(data, normed=True, histtype='stepfilled', range=rngDict[band], bins=100,
                log=False, alpha=0.9)
        ax.set_title(band, fontsize=subPlotFontTitleSize)
        if band > 510:
            ax.set_xlabel(xlbl, fontsize=16)
        if band in [412, 490, 555]:
            ax.set_ylabel(ylbl, fontsize=16)
    if bandNum < 700:
        axs.ravel()[bandNums[bandNum]].hist(dataDict[bandNum].compressed() * 100,
                                            normed=True, histtype='step', linewidth=2,
                                            color='k', range=rngDict[bandNum], bins=100)
    return f, axs
```

```python
# LOAD DATA into dataDict:
# Note that the "baseline" entry does contains the actual baseline data used as 
# a reference; not the differences.
bslDict = dict.fromkeys(['Lt', 'Rrs', 'Geo'])
bslDict['Lt'], bslDict['Rrs'], bslDict['Geo'] = GetL3mProds(os.path.join(mainDir,
                                                                         relPathsDict['baseline']), 
                                                            verbose=False, getLatLon=True)
```

```python
# This will return relative difference only in both Lt and Rrs, 
for k, v in tqdm_notebook(relPathsDict.items()):
    dataDict = dict.fromkeys(['Lt'], ['Rrs'])
    path = os.path.join(mainDir, v)
    dataDict['Lt'], dataDict['Rrs'] = GetDiffs(path, bslDict)
    with open('./Data/L3mDataDictRelDiff_%s.pkl' % k, 'wb') as f:
        pickle.dump(dataDict,file=f)
```

```python
# This will return relative difference only in both Lt and Rrs, 
for k, v in tqdm_notebook(relPathsDict.items()):
    dataDict = dict.fromkeys(['Lt'], ['Rrs'])
    path = os.path.join(mainDir, v)
    dataDict['Lt'], dataDict['Rrs'] = GetDiffs(path, bslDict, computeRel=False)
    with open('./Data/L3mDataDictAbsDiff_%s.pkl' % k, 'wb') as f:
        pickle.dump(dataDict,file=f)
```
