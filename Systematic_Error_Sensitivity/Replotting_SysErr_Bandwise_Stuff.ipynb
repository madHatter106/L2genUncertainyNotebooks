```python
import pickle as pk
import matplotlib.pyplot as pl
import numpy as np
from matplotlib import rcParams
```

```python
rcParams['axes.formatter.limits']=(-2,2)
rcParams['xtick.labelsize'] = 16
rcParams['ytick.labelsize'] = 16
```

```python
% matplotlib notebook
```

```python
exp = 'band1_0p1p'
with open('./Data/L3mDataDictAbsDiff_%s.pkl' % exp, 'rb') as f:
    dataDictAbs = pk.load(f)
```

```python
delrrs443 = dataDictAbs['Rrs'][443].flatten()
mask = np.where((delrrs443<0), True, False)
delrrs443ma = np.ma.array(delrrs443, mask=mask, fill_value=np.NaN)
```

```python
med = np.ma.median(delrrs443ma)
```

06/05/17 - Fred asked me for a histogram of DelRrs for Lt443 perturbation, level 0.1%

Note - It appears the previous absolute difference plots are all wrong and should be discarded

```python
f, ax  = pl.subplots(figsize=(12,8))
rng = (0.25e-4, 1.75e-4)
ax.hist(delrrs443ma.compressed(), bins=50, normed=True, range=rng,
        histtype='stepfilled', alpha=0.8);
ax.hist(delrrs443ma.compressed(), bins=50, normed=True, range=rng,
        histtype='step', color='k', linewidth=2);
ax.set_xlim(rng)
ax.set_xlabel('$\Delta \\ Rrs (sr^{-1})$', fontsize=16);
ax.set_ylabel('$frequency$', fontsize=16);
ylim = ax.get_ylim()
ax.axvline(x=med, color='k', linewidth=3 , ls='--', label='median')
ax.legend(fontsize=16);
f.savefig('/accounts/ekarakoy/Desktop/delRrs443_fromLt443pertubation.png', dpi=300, format='png')
```
